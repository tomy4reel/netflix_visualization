```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Load the dataset
file_path = 'Netflix_shows_movies.csv'
data = pd.read_csv(file_path)
data.head()
```

```python
# Data Cleaning
# Fill missing values in the 'director' and 'cast' columns with 'Unknown'
data['director'].fillna('Unknown', inplace=True)
data['cast'].fillna('Unknown', inplace=True)
```

```python
# Drop rows with any remaining missing values
data.dropna(inplace=True)

# Data Exploration
# Display basic information and summary statistics
print(data.info())
print(data.describe())
```

```python
# Count of unique values in 'type' column (Movies and TV Shows)
type_counts = data['type'].value_counts()
print(type_counts)
```

```python

# Most common genres
genre_counts = data['listed_in'].value_counts().head(10)
print(genre_counts)
```

```python
# Visualization
# Most common genres
plt.figure(figsize=(10, 6))
sns.barplot(y=genre_counts.index, x=genre_counts.values, palette='viridis')
plt.title('Top 10 Most Common Genres on Netflix')
plt.xlabel('Number of Shows/Movies')
plt.ylabel('Genres')
plt.show()
```

```python

# Distribution of Ratings
plt.figure(figsize=(10, 6))
sns.countplot(y='rating', data=data, order=data['rating'].value_counts().index, palette='coolwarm')
plt.title('Distribution of Ratings')
plt.xlabel('Count')
plt.ylabel('Rating')
plt.show()
```

```python
# Save the plot
plt.figure(figsize=(10, 6))
sns.barplot(y=genre_counts.index, x=genre_counts.values, palette='viridis')
plt.title('Top 10 Most Common Genres on Netflix')
plt.xlabel('Number of Shows/Movies')
plt.ylabel('Genres')
plt.savefig('most_common_genres.png')
```
